## 1. 한국영화 리스트 수집

```python
# 패키지 불러오기
from urllib.request import urlopen
from urllib.parse import quote_plus
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd
import xlrd
import re
import os
```

```python
# 최종 데이터 프레임 초기화 
mList = pd.DataFrame()

# 샘플 url
# url: http://movie.naver.com/movie/sdb/browsing/bmovie.nhn?nation=KR&page=1
# 각 페이지당 20 편의 한국영화가 출력되고 있으며, 2017년 11월 23일 현재 총 1345 페이지임

# url 조립
main = "http://movie.naver.com"
sub1 = "/movie/sdb/browsing/bmovie.nhn?nation=KR"
page = 1345

# page를 순환하며 데이터 수집하기 
for i in range(page):
    i += 1
    url = main + sub1 + "&page=" + str(i)
    print(i, url)
    
    # url 접속
    html = urlopen(url)
    
    # html로 읽어오기 
    soup = BeautifulSoup(html, "html.parser")
    
    # 한국영화 데이터 위치 지정
    movies = soup.find("ul", {"class" : "directory_list"})
    movies = movies.findAll("a", href = re.compile("^(/movie/bi/mi/)"))
    
    # 데이터 수집하여 데이터 프레임으로 저장
    for i in range(len(movies)):
        
        # 한국영화 코드
        movieCd = movies[i].attrs["href"].split("=")[1]
        
        # 한국영화 제목
        movieNm = movies[i].get_text().strip()
        
        # 한국영화 링크
        movieLk = main + movies[i].attrs["href"]
        
        # 데이터 프레임으로 저장
        df = pd.DataFrame({"movieCode" : [movieCd],
                           "movieName" : [movieNm],
                           "movieLink" : [movieLk]})
        
        # 최종 데이터 프레임에 추가
        mList = mList.append(df)
```

```python
# 최종 데이터 프레임 확인
mList.head(10)
```

```python
# 데이터 건수 확인
len(mList)
```

```python
# 중복 제거 
mList = mList.drop_duplicates()
len(mList)
```

```python
# 행번호 초기화 후 데이터 확인
mList.reset_index(drop = True, inplace = True)
mList.head(10)
```

```python
# 데이터 저장할 폴더 지정
wdir = "./data/"

# 폴더가 없으면 새로 만들기
if not os.path.exists(wdir):
    os.makedirs(wdir)

# 저장할 파일명 지정
wtime = datetime.today().strftime("%Y%m%d")
fname = wdir + "Korean movie_" + wtime + ".xlsx"

# xlsx 파일로 저장
mList.to_excel(fname, index = False)

# 종료 메시지 출력!
print("It's done!!")
```

## 2. 한국영화 제작년도 수집

```python
# 패키지 불러오기
from urllib.request import urlopen
from urllib.parse import quote_plus
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd
import xlrd
import re
import os
```

```python
# 데이터 폴더 지정
wdir = "./data/"

# 데이터 확인
os.listdir(wdir)
```

```python
# 불러올 파일명 지정 
fname = wdir + "Korean movie_20171123.xlsx"

# xlsx 파일 불러오기
import xlrd
mList = pd.read_excel(fname)

# 영화 리스트 길이 확인
len(mList)
```

```python
# 한국영화 링크만 지정하기 
movies = mList["movieLink"]

# 데이터 확인
movies.head(10)
```

```python
# main url 지정
main = "http://movie.naver.com"

# 최종 데이터 프레임 초기화
mList = pd.DataFrame()

# 수집 진행상황 확인
i = 0
n = len(mList)

# 한국영화 링크 순환하며 제작년도 수집
for movie in movies:
    
    i += 1 
    print(i, "/", n)
    
    # 링크 접속 에러 발생 회피
    try:
        
        # 한국영화 link로 접속 & html로 읽어오기
        html = urlopen(movie)
        soup = BeautifulSoup(html, "html.parser")
        
        # 한국영화 제작년도
        year = soup.find("div", {"class" : "mv_info"})
        year = year.find("strong", {"class" : "h_movie2"}).get_text()
        
        # 4자리 숫자만 
        year = re.compile("\d{4}").findall(year)
        
        # 한국영화 코드
        movieCd = movie.split("=")[1]
        
        # 데이터 프레임으로 저장
        df = pd.DataFrame({"movieCode" : [movieCd],
                           "year" : year})
        
        # 최종 데이터 프레임에 추가 
        mList = mList.append(df)
    
    # 링크 접속 에러 시, 다음 링크로 이동
    except:
        pass
```

```python
# 데이터 건수 확인
len(mList)
```

```python
# 중복 제거 
mList = mList.drop_duplicates()
len(mList)
```

```python
# 행번호 초기화 후 데이터 확인
mList.reset_index(drop = True, inplace = True)
mList.head(10)
```

```python
# 저장할 파일명 지정
wtime = datetime.today().strftime("%Y%m%d")
fname = wdir + "Korean movie year_" + wtime + ".xlsx"

# xlsx 파일로 저장
mList.to_excel(fname, index = False)

# 종료 메시지 출력!
print("It's done!!")
```

## 3. 한국영화 출연 배우 리스트 수집

```python
# 패키지 불러오기
from urllib.request import urlopen
from urllib.parse import quote_plus
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd
import xlrd
import re
import os
```

```python
# 데이터 폴더 지정
wdir = "./data/"

# 데이터 확인
os.listdir(wdir)
```

```python
# 불러올 파일명 지정 
fname = wdir + "Korean movie_20171123.xlsx"

# xlsx 파일 불러오기
mList = pd.read_excel(fname)

# 영화 리스트 길이 확인
len(mList)
```

```python
# 한국영화 링크에서 "배우/제작진" 페이지를 열기 위해 url의 일부를 변경함
# (basic > detail)
movies = mList["movieLink"].str.replace("basic","detail")

# 미리보기
movies.head(10)
```

```python
# main url 지정
main = "http://movie.naver.com"

# 최종 데이터 프레임 초기화
aList = pd.DataFrame()

# 수집 진행상황 확인
i = 0
n = len(mList)

# 한국영화 링크 순환하며 배우/제작진 수집
for movie in movies:
    
    i += 1 
    print(i, "/", n)
    
    # 링크 접속 에러 발생 회피
    try:
        
        # 한국영화 link로 접속 & html로 읽어오기
        html = urlopen(movie)
        soup = BeautifulSoup(html, "html.parser")
        
        # 배우/제작진 리스트 지정
        people = soup.find("div", {"class" : "made_people"})
        people = people.find("ul", {"class" : "lst_people"})
        actors = people.findAll("a", href = re.compile("^(/movie/bi/pi/)"))
        
        # 한국영화 코드
        movieCd = movie.split("=")[1]
        
        # 배우/제작진 리스트에서 개별 데이터 항목 수집
        for j in range(len(actors)):
            
            # 배우 코드
            actorCd = actors[j].attrs["href"].split("=")[1]
            
            # 배우 이름
            actorNm = actors[j].attrs["title"].strip()
            
            # 배우 링크
            actorLk = main + actors[j].attrs["href"]
            
            # 데이터 프레임으로 저장
            df = pd.DataFrame({"movieCode" : [movieCd],
                               "actorCode" : [actorCd],
                               "actorName" : [actorNm],
                               "actorLink" : [actorLk]})
            
            # 최종 데이터 프레임에 추가
            aList = aList.append(df)
    
    # 링크 접속 에러 시, 다음 링크로 이동
    except:
        pass
```

```python
# 데이터 확인 (감독, 배우 동일인인 경우 중복)
aList.head(10)
```

```python
# 중복 제거 (감독, 배우 동일인) 후 건수 확인
aList = aList.drop_duplicates()
len(aList)
```

```python
# 행번호 초기화 후 데이터 확인 
aList.reset_index(drop = True, inplace = True)
aList.head(10)
```

```python
# 저장할 파일명 지정
wtime = datetime.today().strftime("%Y%m%d")
fname = wdir + "Korean movie actors_" + wtime + ".xlsx"

# xlsx 파일로 저장
aList.to_excel(fname, index = False)

# 종료 메시지 출력!
print("It's done!!")
```
